# Q2)Average Number Of Words In Movie Titles Between Year 2000-2005

```python
df
```

```python
df_title = df[(df['release_year']>1999) & (df['release_year']<2006)]
```

```python
word_count = []
for i in range(df_title.shape[0]):
    current_title = str(df['original_title'].values[i])
    current_title_list = current_title.split(' ')
    n=0
    for each_word in current_title_list:
        n = n+1
        word_count.append(n)
```

```python
average_list = sum(word_count) / len(word_count)
```
